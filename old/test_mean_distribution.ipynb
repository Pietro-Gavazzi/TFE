```python
import numpy as np
np.random.seed(42)

sigma = 1
n = 10

normal = []
for i in range (10000):
    mu =  np.random.uniform(-100, 200, 1)
    # print(mu)
    for j in range(1):
        x = np.random.normal(mu, sigma, n)
        xbar = np.mean(x)
        muhat = np.random.normal(xbar, sigma/np.sqrt(n)) # muhat 
        normal+= list(((muhat-mu)*np.sqrt(n/2)))  # 

```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats


# Create histogram
plt.hist(normal, density=True,  bins=30, edgecolor='black', alpha=0.7)

# Generate a normal distribution curve
xmin, xmax = plt.xlim()  # Get the x-axis limits
x = np.linspace(xmin, xmax, 100)  # Generate x values
p = stats.norm.pdf(x,0,1)  # Normal PDF

# Plot the normal curve
plt.plot(x, p, 'r', linewidth=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Show the plot
plt.show()
```

```python
import numpy as np
np.random.seed(42)

sigma = 10

normal = []
for i in range(1):
    beta = np.random.uniform(-10, 2, (10, 1))
    for j in range(1000):
        X = np.random.uniform(-2, 10, (10, 10))
        y = X@beta + np.random.normal(0, sigma, (10, 1))

        betahat = np.linalg.inv(X.T@X)@X.T@y
        variance = np.linalg.inv(X.T@X)*sigma**2
        L = np.linalg.cholesky(variance)

        betahat2 = np.random.multivariate_normal(mean=betahat.T[0],cov=variance)

        unif = np.linalg.inv(L)@(beta.T- betahat2).T/np.sqrt(2)
        normal =normal+list(unif.T[0])

```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats


# Create histogram
plt.hist(normal, density=True,  bins=30, edgecolor='black', alpha=0.7)

# Generate a normal distribution curve
xmin, xmax = plt.xlim()  # Get the x-axis limits
x = np.linspace(xmin, xmax, 100)  # Generate x values
p = stats.norm.pdf(x,0,1)  # Normal PDF

# Plot the normal curve
plt.plot(x, p, 'r', linewidth=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Show the plot
plt.show()
```

```python
import numpy as np
np.random.seed(42)

sigma = 2
n = 11


SSEclassic = 0


normal = []
for i in range(10):
    beta = np.random.uniform(-10, 2, (10, 1))
    for j in range(10):

        X = np.random.uniform(-2, 10, (n, 10))
        y = X@beta + np.random.normal(0, sigma**2, (n, 1))

        variance = np.linalg.pinv(X.T@X)*sigma**2
        L = np.linalg.cholesky(variance)
        betahat = np.linalg.pinv(X.T@X)@X.T@y

        for k in range(10):
            X2 = np.random.uniform(-2, 10, (1, 10))

            ssquared= ((X2@(variance)@X2.T)*sigma**2)[0][0]

            y2 = X2@beta 
            y2hat = X2@betahat

            SSEclassic+=float((y2-y2hat)**2)

            normal+=list(((y2-y2hat)/np.sqrt(ssquared))[0])

# print(normal)
print(np.mean(normal))
print(SSEclassic)
```
