```python

# ------------------------------------------------------------
# DATA GENERATION (your function, unchanged)
# ------------------------------------------------------------
def generate_data(n, d, noise_variance):
    """Generates random Gaussian data with noise."""
    X = random.normal(0, 1, (n, d))
    Xtest = random.normal(0, 1, (n, d))
    w_true = random.normal(0, 1, (d,))
    noise = random.normal(0, 1, (n,))
    y = X @ w_true + noise_variance * noise
    noise2 = random.normal(0, 1, (n,))
    ytest = Xtest @ w_true + noise_variance * noise2
    w_star = np.linalg.pinv(X.T @ X) @ X.T @ y
    norm_factor = np.linalg.norm(w_star)
    w_star = w_star / norm_factor
    y = y / norm_factor
    ytest = ytest / norm_factor
    return (X, y, Xtest, ytest, w_star)


# ------------------------------------------------------------
# BASELINE: standard gradient-descent linear regression
# ------------------------------------------------------------
def train_linear_gd(X, y, lr=1e-2, steps=500):
    n, d = X.shape
    w = np.zeros(d)
    losses = []

    for t in range(steps):
        grad = -(2/n) * X.T @ (y - X @ w)
        w -= lr * grad

        loss = np.mean((y - X @ w)**2)
        losses.append(loss)

    return w, losses


```
